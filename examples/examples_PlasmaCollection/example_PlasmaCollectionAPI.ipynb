```python
from requests import get, post
```

```python
url = 'http://0.0.0.0:8004'
```

```python
result_update_data = get(url + '/plasma/update_data')
result_update_data.json()
```

Use default value (no entry), or explicitly specify stratification by `'site'`. The same for `'population'`.

```python
param = {
    'population': 'itt',  # default
    'strata': ['site']      # default
}
result_get_summary = post(url + '/plasma/get_summary', json = param)
result_get_summary.json()
```

The input and output for distributions are the same + raw data values as tuples. Use `'site'` specific enttries for 1-dim violin plots and `'overall'` for 2-dim scatterplots.

```python
param = {
    'population': 'itt',  # default
    'strata': ['site']      # default
}
result_get_summary = post(url + '/plasma/get_distribution', json = param)
result_get_summary.json()
```
