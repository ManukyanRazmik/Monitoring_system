```python
from requests import get, post
import pandas as pd
```

```python
url = 'http://0.0.0.0:8004'
```

```python
result_update_data = get(url + '/enrolment/update_data')
```

```python
result_update_data.json()
```

```python
show_by = 'week'
site = 'SITE'
stratify_by = ['STRATA1']
by_donor = True
max_weeks = 50
target_sample = 50_000
columns_to_use = ['SITE', 'STRATA1']
donor_level = False
```

```python
metrics_json = {'site': site, 'strata': stratify_by, 'donor_level': by_donor, 'list_of_metrics': ['min', 'max']}
```

```python
result_metrics = post(url + '/enrolment/get_metrics', json=metrics_json)
```

```python
result_metrics.json()
```

```python
bar_chart_json = {'show_by': show_by, 'by_donor': by_donor, 'stratify_by': columns_to_use}
```

```python
bar_chart_json
```

```python
result_bar_chart = post(url + '/enrolment/get_bar_chart', json=bar_chart_json)
```

```python
result_bar_chart.json()
```

```python
projection_steps = pd.DataFrame(result_metrics.json())[columns_to_use + ['min']].rename(columns={'min': 'proj_value'}).to_dict()
```

```python
json_projections = {'max_weeks': 40,
                    'target_sample':50_000,
                    'columns_to_use': columns_to_use,
                    'donor_level': donor_level,
                   'projection_steps': projection_steps}
```

```python
result_projections = post(url + '/enrolment/get_projections', json=json_projections)
```

```python
result_projections.json()
```

```python
json_model_projections = {'max_weeks': 40,
                    'target_sample':50_000,
                    'columns_to_use': columns_to_use,
                    'donor_level': donor_level}
```

```python
result_model_projections = post(url + '/enrolment/get_model_projections', json=json_model_projections)
```
